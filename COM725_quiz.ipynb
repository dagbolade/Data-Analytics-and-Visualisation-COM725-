```python
# import random
import os
from PIL import Image
from IPython.display import display as display_image


# Path to the folder containing the images
image_folder = r"C:\Users\adedi\OneDrive - Solent University\Semester_2\COM725_quiz"

quiz_data = [
    {
        "question": "What is this table called and what is it used for?",
        "image_path": os.path.join(image_folder, "Q_1.PNG"),  
        "options": {
            "A": "This is a confusion matrix and it is useful for evaluating the performance of a classification algorithm.",
            "B": "This is a confusion matrix and it is useful for evaluating the performance of a clustering algorithm.",
            "C": "This is a correlation matrix and it is useful for evaluating the performance of a classification algorithm.",
            "D": "This is a correlation matrix and it is useful for evaluating the performance of a clustering algorithm."
        },
        "correct_answer": "A"
    },
    {
        "question": "Which of the following language should be replaced with the question mark in the below figure?",
        "image_path": os.path.join(image_folder, "Q_2.PNG"),
        "options":{
            "A": "Java",
            "B": "PHP",
            "C": "COBOL",
            "D": "None of the mentioned"
        },
        "correct_answer": "A"
    },
    {
        "question": "Which of the following would be more appropriate to be replaced with question mark in the following figure?",
        "image_path": os.path.join(image_folder, "Q_3.PNG"),
        "options":{
            "A": "Data Analysis",
            "B": "Data science",
            "C": "Descriptive Analytic",
            "D": "None of the mentioned"
        },
        "correct_answer": "B"
    },
    {
        "question": "The following boxplot shows hotel review scores given the traveller type. Select the response that best describes the box plots.",
        "image_path": os.path.join(image_folder, "Q_4.PNG"),
        "options":{
            "A": "There are no outliers in this graphic.",
            "B": "Both the 'Business' and 'Families' traveller types have the biggest distribution of scores - so they have outliers",
            "C": "Three of the five traveller types have outliers as indicated by the black dots lying outside of the black bands of the box plot.",
            "D": "Only the”Solo” traveller type has an outlier because data can only have one outlier"
        },
        "correct_answer": "C"
    },
    {
        "question": "The following chart and small data is the output of a time series forecast for a coffee shop. For each 30-minute time period the forecast for the number of drinks is displayed. The entire timeseries is shown in a line chart. Given that the data is a future forecast, choose the correct analytical method that would produce this type of output.\n What is the correct option?",
        "image_path": os.path.join(image_folder, "Q_5.PNG"),
        "options":{
            "A": "Exploratory",
            "B": "Explanatory",
            "C": "Predictive",
            "D": "None of the mentioned"
        },
        "correct_answer": "C"
    },
    {
        "question": "Here is a histogram of salaries for various jobs in Australia. Each row of the dataset is the average salary for that job, so the counts are counts of jobs.\n Which of the following statements are false? ",
        "image_path": os.path.join(image_folder, "Q_6.PNG"),
        "options":{
            "A": "The histogram is unimodal.",
            "B": "The histogram is bimodal.",
            "C": "The histogram is right-skewed",
            "D": "The histogram is left-skewed",
            "E": "The most popular salary bracket is $40k to $60k."
        },
        "correct_answer": "D"
    },
    {
        "question": "Here are box plots of cigarette consumption per person in the USA from 1985 to 1995 (Alaska and Hawaii are not included). Each observation in the dataset is the average number of packets of cigarette smoked per person in one state in one year. Thus each box plot represents the distribution of 48 data points (because there are 48 US states included in the dataset).\n Which of the following statement is false?",
        "image_path": os.path.join(image_folder, "Q_7.PNG"),
        "options":{
            "A": "The lower quartile number of packets of cigarettes smoked per capita decreased every year from 1987 to 1993.",
            "B": "The inter-quartile range of the number of packets of cigarettes smoked per capita was smallest in 1992.",
            "C": "The median number of packets of cigarettes smoked per capita was below 100 from 1991 onwards.",
            "D": "In 1990, three states were considered to have extreme values in the number of packets of cigarettes smoked per capita."
        },
        "correct_answer": "B"
    },
    {
        "question": "At Hogwarts School of Witchcraft and Wizardry, students often have a lot of homework. The table below indicates the number of hours students studied, and how they performed on an exam. The mean and standard deviation of study hours and exam score have been provided to you in the table below.\n There are multiple ways to calculate a correlation coefficient r (that is, a standardised indicator of the relation between two variables).We can calculate the covariation between two variables (X and Y), and then adjust this by the standard deviation and sample size. Calculate the correlation r between study hours and exam score after filling up the table below.\n Full marks will be awarded upon completion of the table and correct answer.",
        "image_path": os.path.join(image_folder, "Q_8.PNG","Q_9.PNG"),
        "options":{
            "A": "0.27",
            "B": " 0.58 ",
            "C": ". 0.37",
            "D": "0.23"
        },
        "correct_answer": "B"
    },
    {
        "question": "How data science works in general?",
        "options": {
            "A": "Obtain data, analysis, formulate problem, data product",
            "B": "Data product, analysis, formulate problem, obtain data",
            "C": "Formulate problem, obtain data, analysis, data product",
            "D": "Obtain data, formulate problem, analysis, data product"
        },
        "correct_answer": "C",
        "image_path": None
    },
    {
        "question": "Which is the most difficult task usually from the following list?",
        "options": {
            "A": "Causal analysis",
            "B": "Predictive analysis",
            "C": "Inferential analysis",
            "D": "Exploratory analysis"
        },
        "correct_answer": "A",
        "image_path": None
    },
    {
        "question": "Data are categorised as big data when the size Is",
        "options": {
            "A": "10GB - 1 TB",
            "B": "Beyond 1 PB",
            "C": "Up to 10GB",
            "D": "About 1 PB"
        },
        "correct_answer": "B",
        "image_path": None
    },
    {
        "question": "What is the difference between interval/ratio and ordinal variables?",
        "options": {
            "A": "The distance between categories is across the range of interval/ratio data.",
            "B": "Ordinal data can be rank ordered, but interval/ratio data cannot.",
            "C": "Interval/ratio variables contain only two categories.",
            "D": "Ordinal variables have a fixed zero point, whereas interval variables don’t"
        },
        "correct_answer": "D",
        "image_path": None
    },
    {
        "question": "What is the difference between a bar chart and a histogram?",
        "options": {
            "A": "Bar charts represent numbers, whereas histograms represent percentages.",
            "B": "A histogram does not show the entire range of scores in a distribution.",
            "C": "There are no gaps between the bars on a histogram.",
            "D": "Bar charts are circular, whereas histograms are square"
        },
        "correct_answer": "C",
        "image_path": None
    },
    {
        "question": "Similarly, as above, which of the following situations would be most suited to a logistic regression?",
        "options": {
            "A": "Predicting the amount of time, a customer will take to pay for goods they buy",
            "B": "Predicting the number of users arriving at your online store on a given day.",
            "C": "Predicting whether a customer will buy a product that you are selling.",
            "D": "Predicting the total revenue on a given day based on number o customers"
        },
        "correct_answer": "C",
        "image_path": None
    },
    {
        "question": "Which data set, when visualized, would follow a normal distribution?",
        "options": {
            "A": "5, 5, 7, 7, 7, 8, 8, 8, 9",
            "B": "5, 6, 6, 7, 7, 7, 8, 8, 9",
            "C": "5, 5, 5, 5, 5, 6, 7, 8, 9",
            "D": "5, 6, 7, 8, 8, 9, 9, 9, 9"
        },
        "correct_answer": "B",
        "image_path": None
    },
    {
        "question": "Which visualization would you choose to best compare the distribution of sub-groups of a variable, e.g. the distribution of population across continents?",
        "options": {
            "A": "Scatter plot",
            "B": "Bar chart",
            "C": "Box plot",
            "D": "Histogram"
        },
        "correct_answer": "B",
        "image_path": None
    },
    {
        "question": "A study of the relationship between the annual salary of Belgian football coaches and their years of experience as a coach resulted in a linear regression model with the parameters shown. What does an intercept of €23466 mean?",
        "options": {
            "A": "Coaches earn €23466 per year on average irrespective of their experience.",
            "B": "Coaches with no experience earn €23466 per year on average.",
            "C": "Coaches should not expect to earn less than €23466 per year.",
            "D": "None of the above"
        },
        "correct_answer": "B",
        "image_path": None
    },
    {
        "question": "Point out the correct statement.",
        "options": {
            "A": "NAN is the standard missing data marker used in pandas.",
            "B": "Series acts very similarly to an array.",
            "C": "In Python, we use // if we want to comment part of the codes, e.g. //This is a comment.",
            "D": "None of the mentioned"
        },
        "correct_answer": "A",
        "image_path": None
    },
    {
        "question": "Which of the following is an example of raw data?",
        "options": {
            "A": "Original swath files generated from a sonar system",
            "B": "Initial time.series file of temperature values",
            "C": "A real-time GPS-encoded navigation file",
            "D": "All of the mentioned"
        },
        "correct_answer": "D",
        "image_path": None
    },
    {
        "question": "Which of the following is commonly referred to as 'data fishing?",
        "options": {
            "A": "Data bagging",
            "B": "Data booting",
            "C": "Data merging",
            "D": "None of the mentioned"
        },
        "correct_answer": "D",
        "image_path": None
    },
    {
        "question": "Which of the following is one of the key data science skills?",
        "options": {
            "A": "Statistics",
            "B": "Machine Learning",
            "C": "Data Visualization",
            "D": "All of the mentioned"
        },
        "correct_answer": "D",
        "image_path": None
    },
    {
        "question": "Which of the following statements is TRUE?",
        "options": {
            "A": "Outliers should be identified and removed always from a dataset.",
            "B": "Outliers can never be present in the testing dataset.",
            "C": "Outliers are data points that are significantly close to other data points.",
            "D": "The nature of our business problem determines how outliers are used."
        },
        "correct_answer": "D",
        "image_path": None
    },
    {
        "question": "Building an application that takes pictures of all items passing on a conveyor belt and detecting any defects is an example of",
        "options": {
            "A": "Computer Vision",
            "B": "NLP",
            "C": "Pattern recognition",
            "D": "None of the mentioned"
        },
        "correct_answer": "A",
        "image_path": None
    },
    {
        "question": "There are multiple maths functions in Excel that can be used to calculate summary statistics like the minimum, maximum, average, etc. These functions will help make your analysis easier and get a good understanding of your dataset.\n"
                    "Which of the following statements about these functions is false?",
        "options": {
            "A": "The median is the value separating the higher half from the lower half of a data sample.",
            "B": "The SUM() function counts the number of cells that contain a numeric value.",
            "C": "Blank cells in the range of cells that go into the AVERAGE() function will not be included in calculation."
        },
        "correct_answer": "B",
        "image_path": None
    },
    {
        "question": "Point out the wrong statement.",
        "options": {
            "A": "The big volume indeed represents Big Data",
            "B": "The data growth and social media explosion have changed how we look at the data",
            "C": "Big Data is just about lots of data",
            "D": "All of the mentioned"
        },
        "correct_answer": "C",
        "image_path": None
    },
    {
        "question": "Which type of data is generated by POS terminal in a busy supermarket each day?",
        "options": {
            "A": "Source",
            "B": "Processed",
            "C": "Synchronized",
            "D": "All of the mentioned."
        },
        "correct_answer": "B",
        "image_path": None
    }
]


def quiz(limit):
    score = 0
    correct_answers = []
    random.shuffle(quiz_data)

    for i in range(limit):
        print(f"Question {i + 1}: {quiz_data[i]['question']}")
        # Load and display the image inline if image_path is not None
        image_path = quiz_data[i]['image_path']
        if image_path is not None:
            print(f"Image Path: {image_path}")
            image = Image.open(image_path)
            display_image(image)
        else:
            None 
        for option in quiz_data[i]['options']:
            print(f"{option}: {quiz_data[i]['options'][option]}")
        answer = input("Enter your answer: ").upper()
        if answer == quiz_data[i]['correct_answer']:
            score += 1
            correct_answers.append(i + 1)
    return score, correct_answers


# Function to display the score and the correct answers
def display(score, correct_answers):
    print(f"Your score is {score}/{limit}")
    if score == limit:
        print("Well done! You got all the answers correct")
    elif score == 0:
        print("You got all the answers wrong")
        print("The correct answers are:")
        for i in range(limit):
            print(f"Question {i + 1}: {quiz_data[i]['question']}")
            print(f"Correct answer: {quiz_data[i]['correct_answer']}")
            print()
    else:
        print("You got the following answers wrong:")
        for i in range(limit):
            if i + 1 not in correct_answers:
                print(f"Question {i + 1}: {quiz_data[i]['question']}")
                print(
                    f"Correct answer: {quiz_data[i]['correct_answer']}: {quiz_data[i]['options'][quiz_data[i]['correct_answer']]}")
                print()


# Function to allow the user to choose the amount of questions to answer
def choose():
    limit = int(input("How many questions would you like to answer? "))
    if limit > len(quiz_data):
        print(f"Sorry, there are only {len(quiz_data)} questions available")
        limit = int(input("How many questions would you like to answer? "))
    return limit


# Function to allow the user to choose to take the quiz again
def again():
    response = input("Would you like to take the quiz again? (y/n) ")
    if response == "y":
        return True
    elif response == "n":
        print("Thank you for taking the quiz")
    else:
        print("Please enter y or n")
        again()


# code to run the quiz
limit = choose()
score, correct_answers = quiz(limit)
display(score, correct_answers)
while again():
    limit = choose()
    score, correct_answers = quiz(limit)
    display(score, correct_answers)
```
